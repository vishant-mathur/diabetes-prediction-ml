## Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

## Data Collection and Data Preprocessing

```python
df = pd.read_csv("/kaggle/input/docspot/datasets_228_482_diabetes.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
# 0 --> Non-diabetic
# 1 --> Diabetic 
```

```python
df.groupby('Outcome').mean()
```

## Separating Data and Target

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

## Feature Scaling

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

## Train  & Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, stratify = y, random_state = 2)
print(X.shape, X_train.shape, X_test.shape)
```

## Training the Model Model 

```python
model = SVC(kernel='linear')
```

```python
model.fit(X_train,y_train)
```

## Model Evaluation

```python
train_predict = model.predict(X_train)
train_accuracy = accuracy_score(train_predict,y_train)
train_accuracy 
```

```python
test_predict = model.predict(X_test)
test_accuracy = accuracy_score(test_predict,y_test)
test_accuracy 
# Low accuracy due to only less number of rows
```

## Predictive System

```python
input_data = (3,126,88,41,235,39.3,0.704,27)
input_data = np.asarray(input_data)
input_data = input_data.reshape(1,-1)
input_data = scaler.transform(input_data)
predict = model.predict(input_data)
predict[0]
```

```python
if (predict[0] == 1):
    print("Patient is Diabetic")
else:
    print("Patient is Non-Diabetic")
```
